###  **<div class="alert alert-block alert-success"> Supervised Learning</div>**


In supervised learning, we leverage labelled dataset and optimizing methods to find optimal values of parameters. That's why it's called supervised. Here, our labelled data works as a supervisor. 



Problems are grouped into classification and regression. 

* When the output variable is a **category**, such as "True" or "False" then it is known as **classification** problem.

* When the output variable is a **real value (continuous value)**, such as "price of a house" then it is known as **regression** problem. 

**Algorithmns**

- k-nearest neighbor

- linear regression

- logistic regression

- decision trees

- random forest

- neural networks & etc

###  **<div class="alert alert-block alert-success"> Unsupervised Learning</div>**


In Unsupervised Learning, we have input data (X) but no corresponding output labled variable (y). 

Goal of unsupervised learning is to model the distribution of the data in order to learn more about the data. 

Problems are grouped into clustering and regression. 

* When we want to discover **inherent groupings** in the input data it is known as **clustering** problem.

* When we want to discover **rules that describe portions** of the input data it is known as **association** problem. 

* When we are doing **dimensionality reduction** it is also one type of unsupervised learning.

###  **<div class="alert alert-block alert-success"> Semi-Supervised Learning</div>**


Falls between supervised and unsupervised learning, since it uses both labeled and unlabeled data for training.

* typically a small amount of labeled data and a large amount of unlabeled data. 

* systems that use this method are able to considerably improve learning accuracy.

* chosen when the acquired labeled data requires skilled and relevant resources in order to train it or learn from it. Otherwise, acquiring unlabeled data generally doesn’t require additional resources.


###  **<div class="alert alert-block alert-success"> Reinforcement Learning</div>**

In Reinforcement Learning, an agent acts in an environment so as to maximize its rewards. It's kind of trail and error method. Agent learns from past experience and tries to capture best possible knowledge to make accurate decision. For example, an agent which learns to play atari game.

* Q-Learning

* R-Learning

* TD-Learning

# <div class="alert alert-block alert-info"> **BUILDING MODEL**</div>

A machine learning model refers to _**a mathematical configuration (formulation)**_ that is built using previously seen data and is set up to predict new data.

Following is the sequence of steps that are performed iteratively to build a model from scratch.

* Data exploration/ Exploratory Data Analysis (EDA) 
* Data preprocessing
* Splitting data for training and validation
* Preparing a regression/classification model
* Assembling all steps using pipelines
* Fitting (Training) the model
* Running predictions on the model
* Evaluating and visualizing model performance

**<div class="alert alert-block alert-warning"> DATA EXPLORATION</div>**

Statistically or visually exploring to understand what is in a dataset and the characteristics of the data. Biased representation of data results in a skewed model.

**<div class="alert alert-block alert-warning"> DATA PREPROCESSING</div>**

This step is considered to bean important step in the machine learning model building process. This is because the model can only perform well when the data it is trained on is good and well prepared. 
Visit [here](https://scikit-learn.org/stable/modules/preprocessing.html) for more information on data preprocessing.


**<div class="alert alert-block alert-warning"> SPLITTING DATA</div>**

After the data has been preprocessed, the next step is to split the data into parts that can be used to create and train the model, testing and evaluate the model.

**<div class="alert alert-block alert-warning"> ALGORITHM TO BUILD THE MODEL</div>**

Decide on the type of algorithm to use to build your model and initialize required hyperparameters. It is a good practice to try with default hyperparameters and remember to build more than one model.

**<div class="alert alert-block alert-warning"> ASSEMBLE STEPS INTO PIPELINE</div>**

A pipeline is a very convenient process of designing your data processing in a machine learning flow. Data preprocessing is a tedious step that must be applied on data every time before training begins irrespective of the algorithm that will be applied. 

Typical sequence of preprocessing steps that is applied every time before the data modeling begins.

For further reading visit [here](https://www.datanami.com/2018/09/05/how-to-build-a-better-machine-learning-pipeline/). Visit this [site](https://www.ebooks.com/en-mz/book/210083962/building-machine-learning-pipelines/hannes-hapke/) for a copy of the book entitled "Building Machine Learning Pipelines" 

**<div class="alert alert-block alert-warning"> TRAIN THE MODEL</div>**

The final step in creating a model is called modeling. In this step, you basically train your machine learning algorithm. The 98% of data that was split in the splitting data step is used to train the model that was initialized in the previous step.


**<div class="alert alert-block alert-warning"> RUN PREDICTIONS</div>**

After the model is trained, it is ready for some analysis. In this step, the valiadation/test data which was reserved for testing the model is used to run predictions. The predicted output is collected for evaluation against the actual results.

**<div class="alert alert-block alert-warning"> EVALUATE MODEL PERFORMANCE</div>**

This step answers the question, **"how can we determine how bad or good a particular model fits our data?"**

The prediction results obtained in the previous step are compared using what the actual results should have been. 

Evaluation metrics play a key role in the development of a model, as it provides insight to areas that require improvement.

There are different model evaluation metrics for both regression and classification problems.

> **Regression**
 - Mean absolute error
    
 - Root Mean Squared Error (RMSE)
 
 - R-squared 
 
> **Classification**
 - Accuracy
 - Precision
 - Recall
 - F1 Score


**<div class="alert alert-block alert-danger"> NOTE</div>**

<div class="alert alert-block alert-warning"> Until the evaluation outcome is satisfactory, you would repeat all the steps mentioned above from the data preprocessing through evaluating steps by tuning what are called the hyperparameters.</div>

### **LINEAR REGRESSION PROBLEM**

Linear regression is used for regression problems, that is, when target variable is a real value or continuous.

In linear regression, our task is to **establish a relationship** between **target variable(dependent)** and **input variables (independent/features)** by fitting a line**. This line is known as **regression line**. This line can be represented as a linear equation: 

**$y = m * X + b $**. 

 Where, 

- y - target variable, 

- X - input data, 

- m - slope, 

- b - intercept. 

An intercept represents the intersection of regression line with y axis.

**Categorical Variable Analysis**

```python
# count plot for the categorical features
def categorical_feat_countplot(data,list_of_col,figsize = (35,7)):
    fig = plt.figure(figsize=figsize)
    for i,col in enumerate(list_of_col):
        plt.subplot(1,4,i+1)
        sb.countplot(data[col])
        plt.xlabel(col)
        plt.ylabel('Freq')
        plt.title(f'Count Plot of {col.capitalize()}')
        i += 1
    plt.show()  
```

```python
categorical_feat_countplot(student_data,categorical_features)
```

```python
def plot_pie_chart(data,list_of_col,figsize = (25,15)):
    fig = plt.subplots(figsize=figsize)
    for i,col in enumerate(list_of_col):
        plt.subplot(1,4,i+1)
        data[col].value_counts().plot.pie(autopct="%1.1f%%")
        plt.xlabel(col)
        plt.ylabel('')
        plt.title(f'Pie Chart of {col.capitalize()}')
        plt.legend()
        i += 1
    plt.show()  
```

```python
plot_pie_chart(student_data,categorical_features)
```

**Continuous Variable Analysis**

```python
# count plot for the continuous features
def cont_feat_distplot(data,list_of_col,figsize = (35,7)):
    fig = plt.figure(figsize=figsize)
    for i,col in enumerate(list_of_col):
        plt.subplot(1,4,i+1)
        sb.distplot(data[col])
        plt.xlabel(col)
        plt.ylabel('Freq')
        plt.title(f'Distribution Plot of {col.capitalize()}')
        i += 1
    plt.show()  
```

```python
cont_feat_distplot(student_data,continuous_features)
```

```python
# scatter plot for the continuous features against the target variable
def cont_feat_boxplot(data,list_of_col,target,figsize = (35,7)):
    fig = plt.figure(figsize=figsize)
    for i,col in enumerate(list_of_col[:-1]):
        plt.subplot(1,4,i+1)
        sb.scatterplot(data[col],data[target])
        plt.xlabel(col)
        plt.ylabel(target)
        plt.title(f'Distribution Plot of {col.capitalize()}')
        i += 1
    plt.show()  
```

### **<div class="alert alert-block alert-success"> DATA PREPROCESSING</div>**

Now is the time to properly clean our data to ensure that we do not feed our intended model on any garbage (useless information).  

Fortunately for us, we do not have so much to do since the data appears cleaned, the distributions of the continuous variables are somehow normally distributed as well.

**Encode Categorical Features**

```python
student_data_copy = student_data.copy()
student_data_copy.head()
```

```python
student_data_copy = pd.get_dummies(student_data_copy)
student_data_copy.head()
```

**Standardize / Normalize Contiuous Features**

```python
from sklearn.preprocessing import StandardScaler
stan_scaler = StandardScaler()
for col in continuous_features[:-1]:
     feature_array =  student_data_copy[col].values.reshape(-1,1)
     student_data_copy[col] = stan_scaler.fit_transform(feature_array) 
```

```python
student_data_copy.head()
```

### **<div class="alert alert-block alert-success"> SPLITTING DATA</div>**


We will first separate the data into features (independent variables) and target (response / dependent variable) 

```python
features = student_data_copy.drop(columns=['stipend'], axis = 1)
target = student_data_copy['stipend']
print(f'feature: {features.shape}')
print(f'target: {target.shape}')
```

```python
# import a module to help us split the data
from sklearn.model_selection import train_test_split
```

```python
# split data into train and test. set 25% of train data aside to validate the model 
xtrain, xtest, ytrain, ytest = train_test_split(features,target,test_size = 0.25,shuffle = True, random_state = 1)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

### **<div class="alert alert-block alert-success"> ALGORITHM TO BUILD THE MODEL</div>**

Since our target is a continuous variable, then our problem at hand is a regression problem. Hence, we begin our search for the right algorithm with a linear regression.

It is important to note that even the most experienced data scientist can not be perfectly sure of the algorithm that will perform best until it has been tried. Try as many algorithms as you can since it is always difficult to tell from the begiining.

**TRAIN MODEL**

```python
# import from scikit-learn the linear regression module
from sklearn.linear_model import LinearRegression
# instantiate the linear regression model
linear_reg_model = LinearRegression()
# train the model by fitting
linear_reg_model = linear_reg_model.fit(xtrain, ytrain)
```

```python
# Estimated coefficients (weights) for the linear regression problem
print(f'The weights (coefficients) for the linear regression model are: {linear_reg_model.coef_}')
```

```python
# Estimated intercept or offset (bias) for the linear regression problem
print(f'The offset for the for the linear regression model is: {linear_reg_model.intercept_}')
```

### **<div class="alert alert-block alert-success"> RUN PREDICTIONS</div>**

We are going to use the trained linear regression to predict on the 25% of the data that we set aside for testing.

```python
predictions = linear_reg_model.predict(xtest)
predictions[:10]
```

```python
pred_df = pd.DataFrame(data = {'Actual_Stipend':ytest,'Predicted_Stipend':predictions,'Difference':predictions - ytest})
pred_df.head(10)
```

### **<div class="alert alert-block alert-success"> EVALUATE MODEL PERFORMANCE</div>**

<a ><img src = "mtr.png" width = 500, align = "center"></a>

source: https://medium.com/acing-ai/how-to-evaluate-regression-models-d183b4f5853d

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
co_det = r2_score(ytest, predictions)
mae = mean_absolute_error(ytest, predictions)
mse = mean_squared_error(ytest, predictions)
rmse = np.sqrt(mse)
print(f'mean absolute error: {mae}')
print(f'mean square error: {mse}')
print(f'root mean square error: {rmse}')
print(f'coefficient of determination: {co_det}')
```

```python
# visualize with a residual plot
plt.figure(figsize=(10,8))
sb.residplot(predictions,ytest)
plt.xlabel("Actual PRICES")
plt.ylabel("Predicted PRICES")
plt.title("Actual PRICES vs Predicted PRICES")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sb.regplot(ytest,predictions,marker='+',color='r')
plt.xlabel("Actual PRICES")
plt.ylabel("Predicted PRICES")
# plt.legend()
plt.title("ACTUAL PRICES vrs PREDICTED PRICES")
plt.show()
```
